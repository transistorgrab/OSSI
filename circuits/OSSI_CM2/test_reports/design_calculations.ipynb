```python
## a simple schematic to show the components for the calculations afterwards
import schemdraw
import schemdraw.elements as el
with schemdraw.Drawing():
    el.Line().right()
    el.ResistorIEC().down().label("R_1")
    el.Line().right()
    el.ResistorIEC().down().label("R_2")
    el.Line().left()
    el.Ground()
    el.Line().up()
    el.SourceV().up().label("+2V50_REF")
    
```

```python
# calculating tolerances of reference voltages
V_ref = 2.5 ## Volts
V_ref_tol = 0.0015 ## toleranc V_ref
V_ref_max = V_ref * (1 + V_ref_tol)
V_ref_min = V_ref * (1 - V_ref_tol)

R_1 = 1000 ## Ohms
R_1_tol = 0.001 ## tolerance R1
R_1_max = R_1 * (1 + R_1_tol)
R_1_min = R_1 * (1 - R_1_tol)

R_2 = 1000 ## Ohms
R_2_tol = 0.001 ## tolerance R2
R_2_max = R_2 * (1 + R_2_tol)
R_2_min = R_2 * (1 - R_2_tol)

V_ref_half = V_ref / (R_1 + R_2) *  R_2
V_ref_half_min = V_ref_min / (R_1_max + R_2_min) * R_2_min
V_ref_half_max = V_ref_max / (R_1_min + R_2_max) * R_2_max

print(f"Reference Voltage: {V_ref:.3f} V.\n\t Minimum: {V_ref_min:.3f} V.\n\t Maximum: {V_ref_max:.3f} V.")
print(f"0.5 Ã— Reference Voltage : {V_ref_half:.3f} V.\n\t Minimum: {V_ref_half_min:.3f} V.\n\t Maximum: {V_ref_half_max:.3f} V.")
```
